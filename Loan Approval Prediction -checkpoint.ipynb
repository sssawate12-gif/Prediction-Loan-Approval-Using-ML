# Lable Enocoding For Categorical Columns

```python
label=LabelEncoder()
```

```python
for col in df.select_dtypes(include='object').columns:
    df[col]=label.fit_transform(df[col])
```

```python
df.head(5)
```

# Split The Data For Training Purpose

```python
x=df.drop('loan_status',axis=1)
y=df['loan_status']
```

```python
x.head(5)
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
print('x_train.shape:',x_train.shape)
print('x_test.shape:',x_test.shape)
print('y_train.shape:',y_train.shape)
print('y_test.shape:',x_test.shape)
```

# Scaling The Data For Trainin

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

 # Training The Model On Classification Algorithms

```python
models={
    'LogisticRegression':LogisticRegression(),
    'DecisionTreeClassifier':DecisionTreeClassifier(),
    'RandomForestClassifier':RandomForestClassifier(),
    'GradientBoostingClassifier':GradientBoostingClassifier(),
    'XGBClassifier':XGBClassifier()
}
```

```python
from sklearn.metrics import recall_score
```

```python
results=[]
for name,model in models.items():
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)

    print(f"{name}  with accuracy : {accuracy_score(y_test,y_pred)}")
    print("Classification : ",classification_report(y_test,y_pred))
    print("Confusion matrix : ",confusion_matrix(y_test,y_pred))
    print("Precision : ",precision_score(y_test,y_pred))
    print("Recall :",recall_score(y_test,y_pred))
    print("==========================================================")
```

# Calculate The ROC AUC Score & Plot The Curve

```python
y_prob=model.predict_proba(x_test)[:,1]
fpr,tpr,thresholds=roc_curve(y_test,y_prob)

auc_score=roc_auc_score(y_test,y_prob)
print('Roc Auc Curve',auc_score)
```

```python
plt.figure(figsize=(8,6))
plt.plot(fpr,tpr,label=f'AUC={auc_score:.4f}')
plt.plot([0,1],[0,1],'k--',label="Random Model")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Roc Curve')
plt.legend()
plt.show()
```
