### Features Engineering

```python
### Delete Unnecessary Columns Which Carry No Predictive Signals

df.drop(columns=['transaction_id','user_id'],axis=1)
```

```python
### Delete Data Leakage Columns Contain future Output

df.drop(['three_ds_flag','cvv_result','avs_match'],axis=1,inplace=True)
```

```python
### Creating New Features
### New user with high transaction velocity represent elevated fraud risk

df['txn_per_day'] = df['total_transactions_user'] / (df['account_age_days'] + 1)
df['is_new_user'] = (df['account_age_days'] < 30).astype(int)
```

```python
### Create time based features to see the fraud time

df['transaction_time'] = pd.to_datetime(df['transaction_time'])

df['txn_hour'] = df['transaction_time'].dt.hour
```

```python
hourly_fraud_rate = (
    df.groupby('txn_hour')['is_fraud']
      .mean()
)

hourly_fraud_rate.plot(kind='line', marker='o', title='Hourly Fraud Rate')
plt.xlabel('Hour of Day')
plt.ylabel('Fraud Rate')
plt.show()
```

 The graph shows that fraud rates peak during late-night hours, and drop during business
 hours,indicating that time of day is strong behavioral signal for fraud detection

```python
### delete the original column after geting feature out of it 'txn_hour'

df.drop('transaction_time',axis=1,inplace=True)
```

```python
### Categorical Encoding (Lable Encode)
label=LabelEncoder()
for col in df.select_dtypes(include='object').columns:
    df[col]=label.fit_transform(df[col])
```

```python
### Split The Features

x=df.drop(['transaction_id','user_id','is_fraud'],axis=1)
y=df['is_fraud']
```

```python
### Perform train test split for model train & test 
### Target variable is highly imbalance so using stratify 'y' in order to keep proportionate

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,stratify=y,random_state=42)
```

```python
### Print the splitting distribution

print('x_train shape:',x_train.shape)
print('x_test shape:',x_test.shape)
print('y_train shape:',y_train.shape)
print('y_test shape:',y_test.shape)
```

```python
### Scale the data for non tree-based algorithms

scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

## Training The Model With Classification Algorithms

```python
###Since the target variable is highly imbalanced, imbalance-handling techniques
###were applied during model training. For tree-based boosting algorithms,
###`scale_pos_weight` was used to penalize misclassification of the minority
###(fraud) class. For non-tree algorithms, `class_weight='balanced'` was applied
###to ensure the model does not become biased toward the majority class.

neg=(y_train==0).sum()
pos=(y_train==1).sum()

scale_pos_weight=neg / pos
scale_pos_weight
```

```python
models={
    'LogisticRegression':LogisticRegression(class_weight='balanced',random_state=42),
    'DecisionTreeClassifier':DecisionTreeClassifier(class_weight='balanced'),
    'KNeighborsClassifier':KNeighborsClassifier(weights='distance'),
    'RandomForestClassifier':RandomForestClassifier(class_weight='balanced'),
    'XGBClassifier':XGBClassifier(scale_pos_weight=scale_pos_weight,random_state=42),
    'LightGBM':lgb.LGBMClassifier(scale_pos_weight=scale_pos_weight,random_state=42),

}

for name,model in models.items():

    print(f'\n========== {name} ==========')

    model.fit(x_train,y_train)
    
    y_pred=model.predict(x_test)

    print('Accuracy Score')
    print(accuracy_score(y_test,y_pred))
    print('\nClassification Report')
    print(classification_report(y_test,y_pred))
    print('\nConfusion Matrix')
    print(confusion_matrix(y_test,y_pred))
```

Random Forest, XGBoost, and LightGBM showed competitive performance on the
imbalanced fraud dataset. To avoid overfitting and reduce unnecessary model
complexity, the best-performing model was selected and further optimized
through threshold tuning and hyperparameter adjustment.

### Model Tune For Better Performance

```python
### Tune RandomForest Model

rf=RandomForestClassifier(n_estimators=300,
                          max_depth=8,
                          min_samples_split=50,
                          min_samples_leaf=20,
                          class_weight='balanced',
                          random_state=42,
                          n_jobs=-1,)
                          

rf.fit(x_train,y_train)
y_ped=rf.predict_proba(x_test)[:,1]
y_prob=(y_ped > 0.5).astype(int)  ## custom threshold

print('Classification Report')
print(classification_report(y_test,y_prob))
print('\nConfusion Matrix')
print(confusion_matrix(y_test,y_prob))
```

```python
rf=RandomForestClassifier(
    n_estimators=800,
    max_depth=6,
    min_samples_split=30,
    min_samples_leaf=10,
    class_weight='balanced',
    random_state=42,
    n_jobs=-1
)
                          
                          

rf.fit(x_train,y_train)
y_ped=rf.predict_proba(x_test)[:,1]
y_prob=(y_ped > 0.7).astype(int)  ## custom threshold

print('Classification Report')
print(classification_report(y_test,y_prob))
print('\nConfusion Matrix')
print(confusion_matrix(y_test,y_prob))
```

```python
### Find Optimum Thresholds

y_score = rf.predict_proba(x_test)[:,1]

for t in [0.9, 0.8, 0.7, 0.6, 0.5]:
    y_pred = (y_score > t).astype(int)
    p = precision_score(y_test, y_pred)
    r = recall_score(y_test, y_pred)
    print(f"Threshold={t:.1f} | Precision={p:.2f} | Recall={r:.2f}")
```

```python
rf.fit(x_train, y_train)
print("Train AUC:", roc_auc_score(y_train, rf.predict_proba(x_train)[:,1]))
print("Test AUC :", roc_auc_score(y_test, rf.predict_proba(x_test)[:,1]))
```

### Tune Model With XGBClassifiers

```python

xgb=XGBClassifier(
    scale_pos_weight=scale_pos_weight,
    random_state=42,
    
)
xgb.fit(x_train,y_train)
y_pr=xgb.predict_proba(x_test)[:,1]
y_proba=(y_pr > 0.7).astype(int)

print('Classification Report')
print(classification_report(y_test,y_proba))
print('\nConfusion Matrix')
print(confusion_matrix(y_test,y_proba))
```

This XGBoost model shows strong performance on an imbalanced fraud detection problem. It achieves high fraud recall (79%) with acceptable precision (56%), indicating an effective balance between fraud detection and false alerts. The model generalizes well and is suitable for production with threshold tuning based on business cost.

```python
### Precesion Recall Curve

precision,recall,thresholds_pr=precision_recall_curve(y_test,y_pr)
pr_auc=average_precision_score(y_test,y_pr)

plt.figure(figsize=(8,6))
plt.plot(recall,precision,label=f'PR AUC = {pr_auc:.3f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('precision-recall')
plt.legend()
plt.show()
```

This Precision-Recall curve shows the trade-off between catching more positives and introducing
